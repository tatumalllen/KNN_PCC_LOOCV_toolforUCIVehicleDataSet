# Pearson Correlation - Question 1A and 1B

```python
#defining X and Y Values
yValue = df['CLASS']
xValues = ['f0', 'f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'f8', 'f9', 'f10', 'f11', 'f12', 'f13', 'f14', 'f15', 'f16', 'f17', 'f18', 'f19', 'f20', 'f21', 'f22', 'f23', 'f24', 'f25', 'f26', 'f27', 'f28', 'f29', 'f30', 'f31', 'f32', 'f33', 'f34', 'f35']
```

```python
#making an array for the PCC Results
pccResults = []*36
length = len(df[xValues])-1
print(length)
```

```python
#Making PCC Model; this is based on pearson.html
def pearson(x, y):
    sum_sqX = 0
    sum_sqY = 0
    sum_coproduct = 0
    mean_x = 0
    mean_y = 0
    for i in range(length):
        sum_sqX += x[i] * x[i]
        sum_sqY += y[i] * y[i]
        sum_coproduct += x[i] * y[i]
        mean_x += x[i]
        mean_y += y[i]
    mean_x = mean_x / length
    mean_y = mean_y / length
    pop_sd_x = sqrt((sum_sqX / length) - (mean_x * mean_x))
    pop_sd_y = sqrt((sum_sqY / length) - (mean_y * mean_y))
    cov_x_y = (sum_coproduct / length) - (mean_x * mean_y)
    correlation = cov_x_y / (pop_sd_x * pop_sd_y)
    return correlation    
```

```python
#Calculating PCC with scipy 
lenArr = len(xValues)
for i in range(lenArr):
    corr = pearson(df[xValues[i]], yValue)
    pccResults.append(abs(corr))
print(pccResults)
```

```python
#making a 2D array with pccResults and xValues
pccResults = list(zip(pccResults, xValues))
```

```python
#sorting said array by largest to smallest |r| value
len2 = len(pccResults)
pccResults.sort(reverse = True)
for i in range(len2):
    print(pccResults[i])
```

```python
#splitting into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 10)
```

```python
#converting to pandas dataframe to numpy
X_Train = X_train.to_numpy()
X_Test = X_test.to_numpy()
y_Train = y_train.to_numpy()
y_Test = y_test.to_numpy()
```

```python
#knn class
class KNN:
    def __init__(self, k):
        self.k = k
        
    def fit(self, X, y):
        self.X_Train = X
        self.y_Train = y
        
    def distance(self, X1, X2):
        distance = scipy.spatial.distance.euclidean(X1, X2)
    
    def predict(self, X_Test):
        final_output = []
        for i in range(len(X_Test)):
            d = []
            votes = []
            for j in range(len(X_Train)):
                dist = scipy.spatial.distance.euclidean(X_Train[j] , X_Test[i])
                d.append([dist, j])
            d.sort()
            d = d[0:self.k]
            for d, j in d:
                votes.append(y_Train[j])
            ans = Counter(votes).most_common(1)[0][0]
            final_output.append(ans)
            
        return final_output
    
    def score(self, X_Test, y_Test):
        predictions = self.predict(X_Test)
        return (predictions == y_Test).sum() / len(y_Test)
```

```python
#fitting data
knn = KNN(7)
knn.fit(X_Train, y_Train)
```

```python
#prediction
prediction = knn.predict(X_Test)
```

```python
#printing out predicted values
for i in prediction:
    print(i, end= ' ')
```

```python
#knn test score
score = knn.score(X_Test, y_Test)
print(score)
```
